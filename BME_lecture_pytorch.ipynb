### What is PyTorch?

**2. Dynamic Computation Graph:**

**3. Seamless GPU Integration:**
- PyTorch offers seamless integration with CUDA, enabling models to be run on GPUs for accelerated computing. This makes it highly efficient for training large models and processing large datasets.

**4. Extensive Library and Community Support:**
- PyTorch includes a vast range of pre-built layers, tools, and libraries for deep learning. It also benefits from a large community that contributes to its continuous development and expansion.

**5. Research and Production:**
- Initially popular among researchers due to its flexibility and ease of prototyping, PyTorch has also been adopted in production environments. Its ability to switch between eager and graph modes (through TorchScript) makes it versatile for both research and deploying models.

**6. Integration with Python Ecosystem:**
- Being a Python library, PyTorch integrates well with the Python data science stack (like NumPy, SciPy, and Pandas) and visualization libraries, offering a complete ecosystem for AI and machine learning projects.

PyTorch's combination of simplicity, flexibility, and powerful features makes it an excellent choice for both academic research and industrial applications in AI and deep learning.

### Understanding Python Lists, NumPy Arrays, and PyTorch Tensors

**1. Python Lists:**
- Python lists are a collection data type that can hold items of various data types.
- They are dynamic in size and can contain elements of different types (e.g., integers, strings, objects).
- Python lists are versatile but not optimized for numerical computations or element-wise operations.
- Example usage: `my_list = [1, 2, 3, "hello", 4.5]`

**2. NumPy Arrays:**
- NumPy arrays are a core part of the NumPy library, designed for efficient numerical computation.
- They are homogeneous, meaning all elements must be of the same data type.
- NumPy arrays support vectorized operations, allowing for efficient element-wise computations.
- They are generally faster and more compact than Python lists for storing and manipulating large datasets.
- Example usage: `import numpy as np; my_array = np.array([1, 2, 3, 4, 5])`

**3. PyTorch Tensors:**
- PyTorch tensors are similar to NumPy arrays but with additional features that support deep learning and GPU computation.
- Tensors can be used for storing and manipulating data on CPUs or GPUs, which accelerates computations in deep learning models.
- They support automatic differentiation, a key feature in training neural networks.
- PyTorch tensors also allow for seamless GPU acceleration and are optimized for batch processing in machine learning tasks.
- Example usage: `import torch; my_tensor = torch.tensor([1, 2, 3, 4, 5])`

**Key Differences:**
- **Flexibility**: Python lists are more flexible but less efficient for numerical computations.
- **Performance**: NumPy arrays are designed for efficient numerical computation on CPUs.
- **Deep Learning**: PyTorch tensors are optimized for deep learning, supporting GPU computation and automatic differentiation.


### Gradient Descent to Find the Minimum of a Quadratic Function

In this example, we'll use a basic iterative method known as **Gradient Descent** to find the minimum of a quadratic function. The function we are considering is $$ f(x) = (x - 3)^2 $$. This function has a minimum value at $$ x = 3 $$, which we will try to find using gradient descent.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = (x - 3)^2
def f(x):
    return (x - 3)**2

# Generate a range of x values
x = np.linspace(-1, 7, 400)

# Calculate the y values based on the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = (x - 3)^2")
plt.title("Plot of f(x) = (x - 3)^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```


**Gradient Descent** is an optimization algorithm used to minimize functions by iteratively moving towards the steepest descent as defined by the negative of the gradient. In machine learning, this concept is used to minimize loss functions, adjusting model parameters accordingly.

#### Steps of Gradient Descent:
1. **Start with an initial guess**: We begin with a random starting point for our variable $$ x $$.
2. **Compute the gradient**: Calculate the gradient (derivative) of the function at this point.
3. **Update the variable**: Adjust the variable in the direction opposite to the gradient. This is done using the formula: $$ x := x - \alpha \cdot \frac{df}{dx} $$, where $$ \alpha $$ is a small value known as the learning rate.
4. **Repeat**: Repeat the process for a set number of iterations or until the variable converges to a stable value.

In our example, we are iteratively updating the value of `x` and observing how it moves closer to the minimum value of the quadratic function.



This code demonstrates a fundamental concept used in training neural networks, where gradient descent is employed to minimize the loss function, and in doing so, fine-tune the model's parameters.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return (x - 3)**2

def df(x):
    return 2 * (x - 3)

# Gradient Descent
x_start = 0  # Initial guess
learning_rate = 0.1
n_iterations = 15

# Record the progression
x_progression = [x_start]
for _ in range(n_iterations):
    grad = df(x_progression[-1])
    x_next = x_progression[-1] - learning_rate * grad
    x_progression.append(x_next)

# Generate a range of x values
x = np.linspace(-1, 7, 400)
y = f(x)

# Plot the function and the progression
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = (x - 3)^2")
plt.scatter(x_progression, f(np.array(x_progression)), color='red', label='Gradient Descent Progression')
plt.title("Gradient Descent on f(x) = (x - 3)^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

$$ x := x - \alpha \cdot \frac{df}{dx} $$

```python
# Import PyTorch
import torch

# Initial value of x (starting point for gradient descent)
x = torch.tensor([0.0], requires_grad=True)

# Learning rate
learning_rate = 0.1

# Number of iterations for gradient descent
num_iterations = 15

# Gradient Descent
for i in range(num_iterations):
    # Reset the gradients from previous iteration
    if x.grad is not None:
        x.grad.zero_()

    # Compute the function value
    f_x = (x - 3) ** 2

    # Compute gradients
    f_x.backward()

    # Update x by taking a step in the direction opposite to the gradient
    x.data = x.data - learning_rate * x.grad

    # Print the value of x and f(x) after each iteration
    print(f"Iteration {i+1}: x = {x.item()}, f(x) = {f_x.item()}")
```

### Loading and Preprocessing Data

Before training a neural network, we need to load and preprocess our data. In this stage, we'll use the CIFAR-10 dataset, a collection of 60,000 32x32 color images in 10 classes. The dataset will be loaded using PyTorch's `torchvision` library, which provides utilities for loading datasets and transforming them into a format suitable for training.

The main steps include:
1. **Downloading the Dataset**: Using `torchvision.datasets` to automatically download and load the dataset.
2. **Applying Transformations**: Normalizing the images and converting them to PyTorch tensors.
3. **Creating Data Loaders**: Preparing the data for batch-wise training and evaluation.


### What is a Batch in Machine Learning?
In machine learning, particularly in the context of training neural networks, a batch refers to the set of data points (or samples) used in one iteration of training. More specifically, when you train a model using a method like stochastic gradient descent (SGD), you don't typically pass the entire dataset through the network at once. Instead, you divide the dataset into smaller subsets, known as batches.

```python
# Loading and Visualizing CIFAR-10
import torch
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms
import numpy as np

# Transform for the dataset
transform = transforms.Compose([transforms.ToTensor()])

# Load CIFAR-10 dataset
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True)

# Get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# Function to show images
def imshow(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Show images
imshow(torchvision.utils.make_grid(images))
```

### Building the CNN Model

In this section, we'll define our convolutional neural network (CNN) model. We'll use a ResNet architecture, which is known for its effectiveness in image classification tasks. PyTorch provides pre-built models that we can use and customize for our specific dataset.

Key steps:
1. **Defining the Model**: Setting up a ResNet model architecture.
2. **Customizing the Output Layer**: Adjusting the final layer to match the number of classes in the CIFAR-10 dataset.
3. **Model Summary**: Reviewing the model structure to ensure it's ready for training.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # First convolutional layer (input channels = 3, output channels = 64)
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)

        # Second convolutional layer (input channels = 64, output channels = 128)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)

        # Third convolutional layer (input channels = 128, output channels = 256)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1)

        # Fully connected layer
        self.fc = nn.Linear(256, 10)  # Assuming 10 classes for the output

    def forward(self, x):
        # Apply first convolutional layer, ReLU and max pooling
        x = self.pool(F.relu(self.conv1(x)))

        # Apply second convolutional layer, ReLU and max pooling
        x = self.pool(F.relu(self.conv2(x)))

        # Apply third convolutional layer, ReLU and max pooling
        x = self.pool(F.relu(self.conv3(x)))

        # Global Average Pooling
        x = F.adaptive_avg_pool2d(x, (1, 1))

        # Flatten the tensor for the fully connected layer
        x = torch.flatten(x, 1)

        # Apply the fully connected layer
        x = self.fc(x)
        return x

# Create the model instance
model = SimpleCNN()
print(model)
```

```python
# Building and Showing a CNN Model (ResNet)
import torchvision.models as models

# Load a pre-trained ResNet model
resnet = models.resnet18(pretrained=True)
num_ftrs = resnet.fc.in_features
resnet.fc = nn.Linear(num_ftrs, 10)

# Displaying the model structure
print("Model Structure: \n\n", resnet)
```

### Training the CNN Model

Training is the process of feeding our data to the neural network and adjusting the network's weights based on the prediction errors. We'll set up a training loop where the model learns from the CIFAR-10 training images.

Important aspects of training:
1. **Loss Function**: Defining a criterion that measures the difference between the predictions and the actual labels.
2. **Optimizer**: Choosing an optimization algorithm (like SGD or Adam) that updates the model's weights.
3. **Training Loop**: Iterating over the dataset, making predictions, computing loss, and updating the model weights.


```python
# Training the Model
import torch.optim as optim
import torch.nn as nn

# Assuming 'resnet' is the model from Step 5
resnet.train()  # Set the model to training mode

# Loss function
criterion = nn.CrossEntropyLoss()

# Optimizer (e.g., Adam, SGD, etc.)
optimizer = optim.SGD(resnet.parameters(), lr=0.001, momentum=0.9)

# Training loop (Here, we show the structure but won't run it)
for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = resnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print(f"[{epoch + 1}, {i + 1}] loss: {running_loss / 2000:.3f}")
            running_loss = 0.0

print('Finished Training')
```

### Evaluating the Model Performance

After training, we need to evaluate our model's performance on unseen data. This helps us understand how well the model generalizes. We'll use the test set of CIFAR-10 for this purpose.

Evaluation involves:
1. **Disabling Gradient Calculations**: Since we're not training, we don't need to compute gradients.
2. **Testing Loop**: Running the model on the test dataset and computing the accuracy.
3. **Analyzing the Results**: Understanding how well the model performs and identifying areas for improvement.


```python
# Model Evaluation and Inference
resnet.eval()  # Set the model to evaluation mode

# Function to perform inference and visualize results
def visualize_inference(dataloader):
    dataiter = iter(dataloader)
    images, labels = next(dataiter)

    # Print actual labels
    print('Actual labels: ', ' '.join('%5s' % labels[j] for j in range(4)))

    # Perform inference
    outputs = resnet(images)

    # Convert outputs to predicted class
    _, predicted = torch.max(outputs, 1)

    # Print predicted labels
    print('Predicted: ', ' '.join('%5s' % predicted[j] for j in range(4)))

    # Show images
    imshow(torchvision.utils.make_grid(images))

# Visualize inference on some test data
visualize_inference(trainloader)
```

### Saving and Loading Models in PyTorch

In PyTorch, you can save and load models using two main methods: saving the entire model or saving just the model state dictionary. The state dictionary (`state_dict`) is a Python dictionary object that maps each layer to its parameter tensor.

**Saving the Model:**




```python
# Saving the entire model
torch.save(resnet.state_dict(), 'resnet_model.pth')

# Loading the entire model
resnet = models.resnet18(pretrained=False)
model = torch.load_state_dict('resnet_model.pth')
model.eval()  # Set the model to evaluation mode
```
